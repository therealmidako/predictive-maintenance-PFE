```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
df=pd.read_csv('./Datas.csv', encoding='latin-1',sep=";")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop('Modèle OT', axis=1)
df = df.dropna(subset=['date enregistrement'])
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.dropna(subset=['Classification Equipement'])
df = df.dropna(subset=['Service'])
df.isnull().sum()
```

```python
df = df.drop('Description', axis=1)
df = df.drop('Staut Ordre de travail', axis=1)
df = df.drop('Date début', axis=1)
df = df.drop('Date Fin', axis=1)

df.head(20)
```

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate records: {duplicate_count}")
```

```python
df.drop_duplicates(inplace=True)
```

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate records: {duplicate_count}")
```

```python
category_mapping = {
    'Mec': 'MECANIQUE',
    'mec': 'MECANIQUE',
    'MEC': 'MECANIQUE',
    'ELEC': 'ELECTRIQUE',
    'Elec': 'ELECTRIQUE',
    'ALIM': 'ALIM',
    'ENERG': 'ENERG',
    'ELECTRIQUE': 'ELECTRIQUE',
    'GC':'GC',
    'Qual' : 'Qual',
    'PROCESS':'PROCESS',
    'Trait':'Trait',
    '0':'0'
    
}
df['Classification Equipement'] = df['Classification Equipement'].map(category_mapping)
value_counts = df['Classification Equipement'].value_counts()
print(value_counts)
```

```python
df.head(5)
```
